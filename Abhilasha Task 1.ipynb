```python
import numpy as np # linear algebra
import os # accessing directory structure
import pandas as pd
```

```python
# Define the path to the dataset
dataset_path = '/Users/saifalikhan/downloads/KaggleV2-May-2016.csv'

# Load the dataset into a DataFrame
df = pd.read_csv(dataset_path)

# Display the first few rows of the DataFrame
df.head()
```

```python
print("Shape of the dataset:")
df.shape
```

```python
# Get information about the DataFrame
df.info()
```

# Dealing Null Values

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)
df.dropna(inplace=True)
```

# Remove Duplicates

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Check the shape of the DataFrame after removing duplicates
print("Shape of the dataset after removing duplicates:", df.shape)
```

# Standardized text values

```python
# Standardize gender values (example)
df['Gender'] = df['Gender'].str.lower().str.strip()  # Convert to lowercase and strip whitespace
```

# Convert date Formats

```python
# Convert date columns to datetime format without specifying a format
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'], errors='coerce')
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'], errors='coerce')
```

```python
df.head()
```

# Rename column headers

```python
# Rename columns to lowercase and replace spaces with underscores
df.columns = df.columns.str.lower().str.replace(' ', '_')

# Display the new column names
print("Renamed columns:\n", df.columns)
```
